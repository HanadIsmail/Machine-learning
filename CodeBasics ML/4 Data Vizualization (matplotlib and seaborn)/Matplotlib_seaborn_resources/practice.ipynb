```python
import pandas as pd
```

```python
df_sales = pd.read_excel("linechart.xlsx")
df_sales 
```

```python
from matplotlib import pyplot as plt 
```

```python
plt.figure(figsize= (12,4))
plt.plot(df_sales["Quarter"],df_sales["Fridge"],label ="Fridge")
plt.plot(df_sales["Quarter"],df_sales["Dishwasher"],label ="Dishwasher")
plt.plot(df_sales["Quarter"],df_sales["Washing Machine"],label ="Washing Machine")
plt.legend()
plt.show
```

```python
totol_sales = df_sales[["Fridge","Dishwasher", "Washing Machine"]].sum()
totol_sales
```

```python
plt.pie(totol_sales, labels = totol_sales.index,autopct = "%1.1f%%",shadow = True,startangle= 90)
plt.show()
```

```python
df_sales.plot(kind ="bar", x="Quarter")
plt.xticks(rotation= 45)
```
