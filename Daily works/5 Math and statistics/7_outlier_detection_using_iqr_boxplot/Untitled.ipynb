```python
Q1,Q3 = df.height.quantile([0.25,0.75])
Q1,Q3
```

```python
IQR = Q3 -Q1
IQR
```

```python
lower_limit = Q1 - 1.5*IQR
lower_limit
upper_limit = Q3 + 1.5*IQR
```

```python
df[(df.height < lower_limit) |(df.height > upper_limit)]
```

```python
new_df = df[(df.height > lower_limit) &(df.height < upper_limit)]
new_df
```
