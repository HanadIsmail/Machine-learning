```python
def get_lower_upper(data):
    Q1,Q3 = data.Sales.quantile([0.25,0.75])
    IQR = Q3 - Q1
    lower = Q1 - 1.5*IQR
    upper = Q3 + 1.5*IQR
    return lower,upper
```

```python
lower , upper = get_lower_upper(df_apac)
lower , upper
```
